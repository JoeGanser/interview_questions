# Summary of NB2 regression strategy
* STEP 1: Fit the Poisson regression model on the data set. This will give us the vector of fitted rates λ.
* STEP 2: Fit the aux OLS regression model on the data set. This will give us the value of α.
* STEP 3: Use the α from STEP 2 to fit the NB2 regression model to the data set.
* STEP 4: Use the fitted NB2 model to make predictions about expected counts on the test data set.
* STEP 5: Test the goodness-of-fit of the NB2 model.

source: https://timeseriesreasoning.com/contents/negative-binomial-regression-model/

# ETL

```python
import os
import datetime
root = 'bicycle_count/'
files = [os.path.join(root,j) for j in os.listdir(root) if '.xlsx' in j]
df = pd.DataFrame()
for f in files:
    _ = pd.read_excel(f)
    _.columns = _.loc[4]
    _ = _.loc[5:]
    _  = _[['Day','High Temp (°F)','Low Temp (°F)','Precipitation','Brooklyn Bridge']].dropna()
    _['Day'] = pd.to_datetime(_['Day'])
    _['Day'] = _['Day'].apply(lambda x: x.day_name())
    df = pd.concat([_,df],axis=0)
#df.reset_index(inplace=True)

def parse_precipitation(x):
    if x in ['T','S']:
        return 0
    else:
        return x
df['Precipitation'] = df['Precipitation'].apply(lambda x: parse_precipitation(x))
df.index.name = None
df
```
